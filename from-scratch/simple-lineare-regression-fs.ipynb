```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def simple_lineare_regression(X,Y,iterations,learning_rate):
    #l_e=0,0
    a=0.0
    b=0.0
    m=len(Y)
    for i in range(iterations):
        predire=a*X+b
        erreur=predire-Y
        d_a=np.sum(erreur*X)/m
        d_b=np.sum(erreur)/m
        a-=learning_rate*d_a
        b-=learning_rate*d_b
        mse=(np.sum(erreur**2))/(2*m)
        if i%100==0 :
            print(f"loss {mse}")
    return a,b
```

```python
data=pd.read_csv("../datasets/house_prices_dataset.csv")
X_raw = data["area"]
y_raw = data["price"]
X_norm = (X_raw - X_raw.mean()) / X_raw.std()
y_norm = (y_raw - y_raw.mean()) / y_raw.std()
X = X_norm.values
Y = y_norm.values
a, b = simple_lineare_regression(X, Y, 1000, 0.01 )
```

```python
plt.scatter(X,Y)
Y_pred = a * X + b
plt.plot(X,Y_pred,color='red')
plt.show()
```
